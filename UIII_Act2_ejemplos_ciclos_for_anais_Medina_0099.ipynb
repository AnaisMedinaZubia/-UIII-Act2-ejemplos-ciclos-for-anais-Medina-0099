#### Ejemplos de for

```python
# 1. Tabla de multiplicar personalizada
# Este código solicita al usuario un número y genera su tabla de multiplicar del 1 al 10
numero = int(input("Ingrese el número para ver su tabla: "))

# El ciclo for recorre el rango del 1 al 10
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")

print("Programa realizado por Anais Medina NL = 36")
```

```python
# 2. Suma de una cantidad específica de números
# El usuario decide cuántos números desea sumar; el programa pide cada valor y muestra el total acumulado
cantidad = int(input("¿Cuántos números desea sumar? "))
suma_total = 0

# Itera la cantidad de veces que el usuario indicó
for i in range(cantidad):
    valor = float(input(f"Ingrese el número {i+1}: "))
    suma_total += valor

print(f"La suma total es: {suma_total}")
print("Programa realizado por Anais Medina NL = 36")
```

```python
# 3. Contador de vocales en una palabra
# Este ejemplo permite al usuario ingresar una palabra y el ciclo recorre cada letra para contar cuántas son vocales
palabra = input("Ingrese una palabra: ").lower()
contador_vocales = 0

# Recorre cada carácter de la cadena ingresada
for letra in palabra:
    if letra in "aeiou":
        contador_vocales += 1

print(f"La palabra tiene {contador_vocales} vocales.")
print("Programa realizado por Anais Medina NL = 36")
```

```python
# 4. Saludo personalizado a una lista de invitados
# El programa pide al usuario que ingrese nombres separados por comas, los convierte en una lista y genera un saludo para cada uno
entrada = input("Ingrese nombres de invitados separados por coma: ")
lista_nombres = entrada.split(",")

# Itera directamente sobre los elementos de la lista generada
for nombre in lista_nombres:
    # strip() elimina espacios innecesarios alrededor del nombre
    print(f"¡Bienvenido(a), {nombre.strip()}!")
    print("Programa realizado por Anais Medina NL = 36")
```
